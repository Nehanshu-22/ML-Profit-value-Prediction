```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# Loading and reading the .csv file and creating a pandas dataframe "df"
df=pd.read_csv("50_Startups.csv")
df
```

```python
# Data Preprocessing
# Checking for null values
df.isnull().sum()
```

```python
# Creating predictor and target variables X and Y
X=df.drop(['Profit'],axis=1)
Y=np.array(df["Profit"])
Y
```

```python
# Splitting dataset into train set and test set
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=42)
```

```python
# Linear regression model
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,Y_train)
print("Model trained succesfully!!")
Y_pred=reg.predict(X_test)
Y_pred = np.round(reg.predict(X_test))
Y_test = np.round(Y_test)
pd.DataFrame({'Actual Profits': Y_test, 'Predicted Profits': Y_pred,'Difference':Y_test-Y_pred})
```

```python
fig,ax=plt.subplots(figsize=(10,10))
sns.regplot(x=Y_test, y=Y_pred, data=df,scatter_kws={"color": "blue"}, line_kws={"color": "green"})
plt.title("Actual VS Predicted Profits")
plt.xlabel("Actual Profits")
plt.ylabel("Predicted Profits")
```

```python
# Residual Errors
fig,ax=plt.subplots(figsize=(10,10))
plt.style.use('ggplot')
plt.scatter(reg.predict(X_train), reg.predict(X_train) -Y_train,
color = "green", s = 10, label = 'Train data')
plt.scatter(reg.predict(X_test), reg.predict(X_test) - Y_test,
color = "blue", s = 10, label = 'Test data')
plt.hlines(y = 0, xmin = 0, xmax =200000 , linewidth = 2,color="r")
plt.legend(loc = 'upper right')
plt.title("Residual Errors")
plt.show()
```

```python
# Regression metrics for the Linear Regression model
#1. Mean Absolute error(MAE)
from sklearn.metrics import mean_absolute_error as mae
train_predict = reg.predict(X_train)
k = mae(train_predict, Y_train)
print('Training Mean Absolute Error', k )

test_predict = reg.predict(X_test)
mae_lin_reg= mae(test_predict,Y_test)
print('Test Mean Absolute Error    ', mae_lin_reg )

#2. Mean absolute percentage error(MAPE)
from sklearn.metrics import mean_absolute_percentage_error as mape
train_predict = reg.predict(X_train)
k = mape(train_predict, Y_train)
print('Training Mean Absolute Percentage Error:', k )

test_predict = reg.predict(X_test)
k = mape(test_predict,Y_test)
print('Test Mean Absolute Percentage Error:', k )

#3. Mean Squared Error(MSE)
from sklearn.metrics import mean_squared_error as mse
train_predict = reg.predict(X_train)
k= mse(train_predict, Y_train)
print('Training Mean Squared Error:', k )

test_predict = reg.predict(X_test)
mse_lin_reg = mse(test_predict,Y_test)
print('Test Mean Squared Error:', mse_lin_reg )

#4. Root mean squared error (RMSE)
from sklearn.metrics import mean_squared_error
from math import sqrt

rmse_lin_reg = sqrt(mean_squared_error(Y_test, Y_pred))
print("Root Mean squared error:",rmse_lin_reg)

#5. R-squared
from sklearn.metrics import r2_score
r2_lin_reg = r2_score(Y_test,Y_pred)
print("R2 score:",r2_lin_reg)
#6. Adjusted  R-squared

Adj_r2_lin_reg = 1 - (1-r2_score(Y_test, Y_pred)) * (len(Y)-1)/(len(Y)-X.shape[1]-1)
print("Adjusted R-squared:",Adj_r2_lin_reg)
#7.#Score

Test=reg.score(X_test,Y_test)
Train=reg.score(X_train,Y_train)
print("Model test score is",Test)
print("Model train score is",Train)
```

```python
#Ridge Regression Model
from sklearn.linear_model import Ridge, RidgeCV, Lasso
ridgeReg = Ridge(alpha=0.01)
ridgeReg.fit(X_train,Y_train)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
folds=KFold(n_splits=5,shuffle=True,random_state=42)
# hyperparameters range
params={'alpha':[0.001,0.01,0.1,0.2,0.5,0.9,1.0,5.0,10.0,20.0]}
model=Ridge()
model_cv=GridSearchCV(estimator=model,param_grid=params,scoring="r2",cv=folds,return_train_score=True,verbose=1)
model_cv.fit(X_train,Y_train)

model_cv.best_params_
```
